```python
ia = imdb.IMDb()
for i in range(120737,167262):
    try:
        movie=ia.get_movie(i)
        temp1=[]
        temp2=[]
        for person in movie['cast']:
            temp1.append(person.personID)
            temp2.append(person['name'])
        
        movie_date.append(movie['year'])
        movie_name.append(movie['title'])
        movie_castId.append(temp1)
        movie_castName.append(temp2)
        movie_id.append(i)
    except:
        continue
```

```python
tempdict={}

for i in range(len(movie_id)):
    id=i+1+31567
    tempdict.update({
        id:{
        "MovieID":movie_id[i],
        "Name":movie_name[i],
        "Year":movie_date[i],
        "CastId":movie_castId[i],
        "CastName":movie_castName[i]
        }})
```

```python
import json

json_object = json.dumps(tempdict, indent = 4,ensure_ascii=False)
with open ("movie.json","a",encoding='utf-8') as f:
    f.write(json_object)
```
